# Q1. Identify and rectify the eight prominent data quality issues present in the given datasets.    

#### Issue 1 : Duplicate data
#### YoutubeLikeDislikes column in the dataset1 is the sum of the values in two other columns in the same dataset. 
#### It is also not present in the other dataset. Hence removed it from dataset1.

#### Issue 2 :  Invalid data
#### ReleaseTime column in dataset1 has an invalid value (44). Corrected that to 4.

#### Issue 3 : Inconsitent data
#### Release_Time in dataset2 has different values compared to that of ReleaseTime in dataset1. Modified the dataset2
#### as per the mapping [Long Weekend (3), Festive Season (1), Holiday Season (2), and Normal (4)] to make the datasets
#### consistent.

#### Issue 4 : Noise
####  Genre field has the string 'Genre' appended or prepended for all the data values. Similarly, Budget field had 'Cr' added or pretexted for few data values. 

#### Issue 5 : Missing Data
#### YoutubeLikes and YoutubeDislikes fields in dataset1 has 4 null values. Removed those 4 rows from the dataset.

#### Issue 6 : Missing Data
#### The values of Release_Date is missing for 4 records in the dataset1.
