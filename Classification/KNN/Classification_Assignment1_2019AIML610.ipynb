### Seperating the Outcome column from the dataset

```python
X = df.drop('Outcome', axis = 1).values
y = df['Outcome'].values
```

### Scaling the data

```python
from sklearn.preprocessing import StandardScaler
standard_scalar = StandardScaler()
X =  pd.DataFrame(standard_scalar.fit_transform(X),
        columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42, stratify = y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
neighbors = np.arange(1,20)

train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
model_accuracy = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors = k)
    
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test)
    model_accuracy[i] = metrics.accuracy_score(y_test, y_pred)
print(train_accuracy)
print('-----------------')
print(test_accuracy)
print('-----------------')
print(model_accuracy)
```

```python
plt.title('KNN Graph varying number of neighbors')
plt.plot(neighbors, train_accuracy, label='Training Accuracy')
plt.plot(neighbors, test_accuracy, label='Testing Accuracy')
plt.plot(neighbors, model_accuracy, label='Model Accuracy')
plt.legend()
plt.xlabel('No of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

### Best value of K seems to be 11 from the above graph. Hence creating the final model with value of K as 11

```python
knn = KNeighborsClassifier(11)

knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
train_accuracy[i] = knn.score(X_train, y_train)
test_accuracy[i] = knn.score(X_test, y_test)
model_accuracy[i] = metrics.accuracy_score(y_test, y_pred)

print(train_accuracy)
print('-----------------')
print(test_accuracy)
print('-----------------')
print(model_accuracy)
```

```python
#import confusion_matrix
from sklearn.metrics import confusion_matrix
#let us get the predictions using the classifier we had fit above
y_pred = knn.predict(X_test)
cnf_matrix = confusion_matrix(y_test,y_pred)
pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
```
